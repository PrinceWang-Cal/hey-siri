```python

def graph_to_matrix(input_graph, weight_low, weight_high, randomize):
    
    #compute number of nodes in the graph
    node_count = input_graph.number_of_nodes()
    
    #convert graph into adjacency list formate
    graph_adjlst = nx.generate_adjlist(input_graph)
    
    #Create a nxn matrix of zero entries where n is node count
    output = np.zeros((node_count, node_count))
    
    
    for line in graph_adjlst:
        splitted_line = line.split(" ")
        row_num = int(splitted_line[0])
        output.itemset((row_num, row_num), 0)
        splitted_line = splitted_line[1:]
        for col_num in splitted_line:
            col_num = int(col_num)
            
            if randomize:
                weight = np.random.randint(low = weight_low, high = weight_high)
            else:
                weight = 1
                
            output.itemset((row_num, col_num), weight)
            output.itemset((col_num, row_num), weight)
    return output
```

```python
new_graph = graph_to_matrix(G, 5, 10, True)
type(new_graph)
```
